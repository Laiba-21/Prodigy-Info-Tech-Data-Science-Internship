```python
import matplotlib.pyplot as plt


# Sample data – replace these with your actual data

Pak_Country = total_Population[total_Population['Country Name'] == 'Pakistan']

Years = total_Population.columns[4:]

# Extract population values from columns 4 to 10 (as a 1D list)
Population = Pak_Country[Years].values.flatten()  # or .tolist()

# Create the bar chart
plt.figure(figsize=(20, 10))
plt.bar(Years,Population)

# Add labels and title
plt.xlabel("Pakistan Country")
plt.ylabel("Population (in millions)")
plt.title("TOTAL Population Growth Through the Years")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Combine them into one DataFrame
df_combined = pd.concat([female_Population, male_Population], ignore_index=True)

# Save the combined file (optional)
df_combined.to_excel("C://Users//Dell//Downloads//World Population Prospects//combined_population.xlsx", index=False)

print("Combined dataset saved as 'combined_population.xlsx'")
print(df_combined)
```

```python
df_combined['Indicator Name'] = df_combined['Indicator Name'].replace({
    'Population, female': 'Female',
    'Population, male': 'Male'
})

print(df_combined)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter for one country if needed
country_name = "Pakistan"
df_country = df_combined[df_combined['Country Name'] == country_name]

# Convert wide → long
df_long = df_country.melt(
    id_vars=['Country Name', 'Indicator Name'],  # Indicator Name = Male/Female
    var_name='Year',
    value_name='Population'
)

# Ensure Year is numeric
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')

# Pivot so we have columns: Year, Female, Male
df_pivot = df_long.pivot_table(
    index='Year',
    columns='Indicator Name',
    values='Population',
    aggfunc='sum'
).reset_index()

# Plot stacked bar chart
plt.figure(figsize=(14, 6))
plt.bar(df_pivot['Year'], df_pivot['Female'], label='Female')
plt.bar(df_pivot['Year'], df_pivot['Male'], bottom=df_pivot['Female'], label='Male')

plt.title(f"{country_name} - Population by Gender (1960–2024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
